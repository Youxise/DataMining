```python
import matplotlib.pyplot as plt
```

```python
def load(path):

  with open(path, 'r') as file:
    temp = file.read()
  # separate rows
  temp = temp.split("\n")
  data = []
  # separate columns
  for elt in temp:
    elt = elt.split(",")
    data.append(elt)
  return data
```

1. Scatter plot generation

```python
from itertools import combinations

def dispersion(data):
    # init variables
    x = []
    y = []
    attributes = [0, 1, 2, 3]
    all_combinations = []
    # getting all combinations
    for r in range(2, 3):
        all_combinations.extend(combinations(attributes, r))
    # getting attributes' values
    for combination in all_combinations:
        attribute1 = combination[0]
        attribute2 = combination[1]
        print("( Attribute", attribute1, "; Attribute", attribute2, ")")
        for elt in data:
            x.append(float(elt[attribute1]))
            y.append(float(elt[attribute2]))
        # display
        plt.scatter(x, y)
        plt.title('Diagramme de dispersion')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.show()
        print("--------------------------------------------------------------------")
```

```python
dispersion(data)
```

- Encoding the target column in order to be able to generate its histogram

```python
def encoding(data):
  # init variables & getting all target values
  liste = []
  for elt in data:
    liste.append(elt[-1])
  # getting distinct values
  liste = list(set(liste))
  # encoding
  dictC = {value: index for index, value in enumerate(liste)}
  # replacing values in dataset
  for elt in data:
    elt[-1] = str(dictC[elt[-1]])

  return data
```

```python
data = encoding(data)
data
```

2. Histogram generation

```python
def histogramme(data, i):
    if i == 4:
      # encoding target values
      data = encoding(data)
    # init variables
    x = []
    # getting attributes' values
    for elt in data:
        x.append(int(float(elt[i])))
    # display
    plt.hist(x, bins=max(x)+1, edgecolor='black')
```

```python
histogramme(data, 4)
```

```python
def resume(data, attribut):

  resume = []
  # out of range attribute
  if attribut >= len(data[0]):
    return resume
  # init variables
  sortedElts = []
  # sorting elements
  for elt in data:
    # alphabetic values case --> return null
    if any(char.isalpha() for char in elt[attribut]):
      return resume
    numb = float(elt[attribut])
    sortedElts.append(numb)
  sortedElts.sort()
  # calculating quartiles
  resume.append(min(sortedElts))                      # Q0 = min
  resume.append(sortedElts[int(len(sortedElts)/4)])   # Q1
  resume.append(sortedElts[int(len(sortedElts)/2)])   # Q2 = mediane
  resume.append(sortedElts[int(len(sortedElts)*3/4)]) # Q3
  resume.append(max(sortedElts))                      # Q4 = max
  return resume
```

3. Boxplot (with & without outliers)

```python
def boiteAMoustache(data, attribut, box):
    # getting quartiles
    quartiles = resume(data, attribut)
    # init variables
    IQR = quartiles[3] - quartiles[1]
    x = []
    outliers = []
    # getting attributes' values
    for elt in data:
        numb = float(elt[attribut])
        # separating outliers and normal values
        if numb < (quartiles[1] - 1.5*IQR) or numb > (quartiles[3] + 1.5*IQR):
            outliers.append(numb)
        else:
            x.append(numb)
    # w/o outliers
    if box == 0:
        plt.boxplot(x)
    else:
        x = x + outliers
        plt.boxplot(x)
```

```python
boiteAMoustache(data, 1, 1)
```
